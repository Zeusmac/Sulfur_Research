```python
import ROOT
import math
import numpy as np

ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetOptFit(1)

def format_number(x):
    ax = abs(x)
    if ax > 1e4 or (ax < 1e-3 and ax > 0):
        return f"{x:.6e}"
    else:
        return f"{x:.6f}"
```

```python
gCfg = None

def total_model(x, par):
    t = x[0]
    if t < 0:
        return par[7]

    use_beta = gCfg.use_beta
    use_bn   = gCfg.use_bn
    use_2n   = gCfg.use_2n

    lambda_p, lambda_d, lambda_bn, lambda_gd = par[0], par[1], par[2], par[3]
    lambda_2n_d, lambda_bn_gd = par[4], par[5]
    N0, bg = par[6], par[7]

    eff_p, eff_d, eff_bn = par[8], par[9], par[10]
    eff_gd, eff_2n_d, eff_bn_gd = par[11], par[12], par[13]

    Pn, P2n = par[14], par[15]

    # clamp
    Pn = max(0, min(1, Pn))
    P2n = max(0, min(1, P2n))

    raw_beta = max(0, 1 - Pn - P2n)

    total_branch = 0
    if use_beta: total_branch += raw_beta
    if use_bn: total_branch += Pn
    if use_2n: total_branch += P2n

    if total_branch <= 0:
        return 1e30

    branch_d = raw_beta / total_branch if use_beta else 0
    branch_bn = Pn / total_branch if use_bn else 0
    branch_2n = P2n / total_branch if use_2n else 0

    rate = 0

    # Parent
    Np = N0 * math.exp(-lambda_p * t)
    rate += eff_p * lambda_p * Np

    # Daughter
    if use_beta:
        denom = lambda_d - lambda_p or 1e-12
        Nd = branch_d * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_d*t))
        rate += eff_d * lambda_d * Nd

    # Beta-n
    if use_bn:
        denom = lambda_bn - lambda_p or 1e-12
        Nbn = branch_bn * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_bn*t))
        rate += eff_bn * lambda_bn * Nbn

    # Beta-2n
    if use_2n:
        denom = lambda_2n_d - lambda_p or 1e-12
        N2n = branch_2n * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_2n_d*t))
        rate += eff_2n_d * lambda_2n_d * N2n

    rate += bg

    return max(rate, 0)
```

```python
class FitConfig:
    def __init__(self):
        self.filename = ""
        self.histname = ""
        self.xmin = 0
        self.xmax = 0
        self.rebin = 1
        self.outfile = ""

        self.names = []
        self.init = []
        self.bounds = []

        self.use_beta = 1
        self.use_bn = 1
        self.use_2n = 1

def get_pars(cfg, filename):
    with open(filename) as f:
        for line in f:
            if not line.strip() or line.startswith("#"):
                continue

            tokens = line.split()
            key = tokens[0]

            if key == "FILE":
                cfg.filename = tokens[1]
            elif key == "HIST":
                cfg.histname = tokens[1]
            elif key == "RANGE":
                cfg.xmin = float(tokens[1])
                cfg.xmax = float(tokens[2])
            elif key == "REBIN":
                cfg.rebin = int(tokens[1])
            elif key == "USE_BETA":
                cfg.use_beta = int(tokens[1])
            elif key == "USE_BN":
                cfg.use_bn = int(tokens[1])
            elif key == "USE_2N":
                cfg.use_2n = int(tokens[1])
            elif key == "OUTPUT":
                cfg.outfile = tokens[1]
            else:
                name = key
                init, lo, hi = map(float, tokens[1:])
                cfg.names.append(name)
                cfg.init.append(init)
                cfg.bounds.append((lo, hi))
```

```python
gCfg = None

def total_model(x, par):
    t = x[0]
    if t < 0:
        return par[7]

    use_beta = gCfg.use_beta
    use_bn   = gCfg.use_bn
    use_2n   = gCfg.use_2n

    lambda_p, lambda_d, lambda_bn, lambda_gd = par[0], par[1], par[2], par[3]
    lambda_2n_d, lambda_bn_gd = par[4], par[5]
    N0, bg = par[6], par[7]

    eff_p, eff_d, eff_bn = par[8], par[9], par[10]
    eff_gd, eff_2n_d, eff_bn_gd = par[11], par[12], par[13]

    Pn, P2n = par[14], par[15]

    # clamp
    Pn = max(0, min(1, Pn))
    P2n = max(0, min(1, P2n))

    raw_beta = max(0, 1 - Pn - P2n)

    total_branch = 0
    if use_beta: total_branch += raw_beta
    if use_bn: total_branch += Pn
    if use_2n: total_branch += P2n

    if total_branch <= 0:
        return 1e30

    branch_d = raw_beta / total_branch if use_beta else 0
    branch_bn = Pn / total_branch if use_bn else 0
    branch_2n = P2n / total_branch if use_2n else 0

    rate = 0

    # Parent
    Np = N0 * math.exp(-lambda_p * t)
    rate += eff_p * lambda_p * Np

    # Daughter
    if use_beta:
        denom = lambda_d - lambda_p or 1e-12
        Nd = branch_d * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_d*t))
        rate += eff_d * lambda_d * Nd

    # Beta-n
    if use_bn:
        denom = lambda_bn - lambda_p or 1e-12
        Nbn = branch_bn * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_bn*t))
        rate += eff_bn * lambda_bn * Nbn

    # Beta-2n
    if use_2n:
        denom = lambda_2n_d - lambda_p or 1e-12
        N2n = branch_2n * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_2n_d*t))
        rate += eff_2n_d * lambda_2n_d * N2n

    rate += bg

    return max(rate, 0)
```

```python
def beta_n_daughter(x, par):
    t = x[0]
    lambda_p, lambda_bn = par[0], par[2]
    N0, eff_bn, Pn, bg = par[6], par[10], par[14], par[7]

    denom = lambda_bn - lambda_p or 1e-12
    Nbn = Pn * N0 * (lambda_p/denom) * (math.exp(-lambda_p*t) - math.exp(-lambda_bn*t))

    return eff_bn * lambda_bn * Nbn + bg


def beta_n_granddaughter(x, par):
    t = x[0]
    lambda_p, lambda_bn, lambda_bn_gd = par[0], par[2], par[5]
    N0, eff_bn_gd, Pn, bg = par[6], par[13], par[14], par[7]

    denom1 = (lambda_bn-lambda_p)*(lambda_bn_gd-lambda_p) or 1e-20
    denom2 = (lambda_p-lambda_bn)*(lambda_bn_gd-lambda_bn) or 1e-20
    denom3 = (lambda_p-lambda_bn_gd)*(lambda_bn-lambda_bn_gd) or 1e-20

    Nbgd = Pn * N0 * lambda_p * lambda_bn * (
        math.exp(-lambda_p*t)/denom1 +
        math.exp(-lambda_bn*t)/denom2 +
        math.exp(-lambda_bn_gd*t)/denom3
    )

    return eff_bn_gd * lambda_bn_gd * Nbgd + bg
```

```python
cfg = FitConfig()
get_pars(cfg, "config.txt")
gCfg = cfg

f = ROOT.TFile.Open(cfg.filename)
h = f.Get(cfg.histname)

h.Rebin(cfg.rebin)

fit = ROOT.TF1("fit", total_model, cfg.xmin, cfg.xmax, len(cfg.init))

for i in range(len(cfg.init)):
    fit.SetParName(i, cfg.names[i])
    fit.SetParameter(i, cfg.init[i])
    fit.SetParLimits(i, cfg.bounds[i][0], cfg.bounds[i][1])

result = h.Fit(fit, "S R M E")
```

```python
c = ROOT.TCanvas("c", "Fit", 800, 600)

# Components
f_bn_d = ROOT.TF1("bn_d", beta_n_daughter, cfg.xmin, cfg.xmax, len(cfg.init))
f_bn_gd = ROOT.TF1("bn_gd", beta_n_granddaughter, cfg.xmin, cfg.xmax, len(cfg.init))

for i in range(len(cfg.init)):
    val = fit.GetParameter(i)
    f_bn_d.SetParameter(i, val)
    f_bn_gd.SetParameter(i, val)

f_bn_d.SetLineColor(ROOT.kBlue)
f_bn_gd.SetLineColor(ROOT.kGreen+2)

h.Draw("E")
fit.Draw("same")
f_bn_d.Draw("same")
f_bn_gd.Draw("same")

leg = ROOT.TLegend(0.6,0.6,0.9,0.9)
leg.AddEntry(h, "Data", "lep")
leg.AddEntry(fit, "Total Fit", "l")
leg.AddEntry(f_bn_d, "Beta-n Daughter + bg", "l")
leg.AddEntry(f_bn_gd, "Beta-n Granddaughter + bg", "l")
leg.Draw()

c.SaveAs("fit.root")
```
